```python
# Normal python
import time
a = [i for i in range(10000000)]
b = [i for i in range(10000000,20000000)]
c =[]

start = time.time()
for i in range(len(a)):  #need to list to calculate one by one
    c.append(a[i]*b[i])
print(time.time() - start)
```

```python
#using same thing Numpy
import numpy as np

a = np.arange(10000000)
b = np.arange(10000000,20000000)

start = time.time()
c = a + b
print(time.time() - start)
```

```python
a[:,[0,2,3]]  # for the column
```

```python
# Boolean Indexinng

a = np.random.randint(1,100,30).reshape(6,5)
a
```

```python
# find all number greater than 50

a[a >= 50]
```

```python
# find out even numbers


a[a % 2== 0]
```

```python
# find all number greater than 50 and even

a[(a > 50) & (a % 2 == 0)]
```

```python
# find all numbers that not dividable by 7

a[a % 7 != 0]

a[~(a % 7 == 0)]
# both are useful for this condition
```

# Broadcasting

The term broadcasting describe how Numpy treats arrays with different shapes during aruthmentic operations.

The smaller array is "broadcast" across the larger array to so that they have commplatiable shapes.



```python
# same shape
a = np.arange(6).reshape(3,2)
b = np.arange(6,12).reshape(3,2)

print(a)
print(b)

print("\n",a + b)
```

```python
# diff shape

a = np.arange(6).reshape(2,3)
b = np.arange(3).reshape(1,3)

print(a)
print(b)

print("\n",a + b)
```

```python
#Some practise of broadcasting

a = np.arange(12).reshape(4,3)
b = np.arange(3).reshape(3)

print(a)
print(b)

print("\n", a + b)
```

```python

a = np.arange(3).reshape(1,3)
b = np.arange(3).reshape(3,1)

print(a)
print(b)

print("\n", a + b)
```

```python

a = np.arange(4).reshape(4,1)
b = np.arange(3).reshape(1,3)

print(a)
print(b)

print("\n", a + b)
```

```python

a = np.arange(1)
b = np.arange(4).reshape(2,2)

print(a)
print(b)

print("\n", a + b)
```

```python
#This don't work because we need 1 in on of the range of array
a = np.arange(12).reshape(4,3)
b = np.arange(12).reshape(3,4)

print(a)
print(b)

print("\n", a + b)
```

##Working with mathematical formulas

```python
#sigmoid

def sigmoid(array):
  return 1/(1 + np.exp(-array))

a = np.arange(10)

sigmoid(a)
```

```python
#mean squered error

actual = np.random.randint(1,20,10)
predict = np.random.randint(1,20,10)

def mse(actual,predict):
  return np.mean((actual - predict)**2)

mse(actual,predict)
```
