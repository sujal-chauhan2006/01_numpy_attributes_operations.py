# Why we use numpy

```python
# Normal python
import time
a = [i for i in range(10000000)]
b = [i for i in range(10000000,20000000)]
c =[]

start = time.time()
for i in range(len(a)):  #need to list to calculate one by one
    c.append(a[i]*b[i])
print(time.time() - start)
```

```python
#using same thing Numpy
import numpy as np

a = np.arange(10000000)
b = np.arange(10000000,20000000)

start = time.time()
c = a + b
print(time.time() - start)
```

```python
#Differnece is shocking,numpy faster than normal python by 10 times
1.9060945510864258/0.17719292640686035

```
